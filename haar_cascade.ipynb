```python
cascade_path = 'E:\\Computer-Vision\\OCR_project\\data\\models\\haarcascade_frontalface_default.xml'
image_path = 'E:\\Computer-Vision\\OCR_project\\data\\raw_images\\nepali-passport-1.jpg'
```

```python
import cv2
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread(image_path)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY, )
plt.imshow(gray_image, cmap = 'gray')
plt.show()
```

```python
face_cascade = cv2.CascadeClassifier(cascade_path)
faces = face_cascade.detectMultiScale(gray_image)
```

```python
# find the largest face
max_area = 0
largest_face = None 

for x,y,w,h in faces:
    area = w*h 
    if area > max_area:
        max_area = area 
        largest_face = (x,y,w,h)
    
if largest_face is not None:
    (x, y, w, h) = largest_face

# extract the largest face 
new_w = int(w * 1.50)
new_h = int(h * 1.50)
# Calculate new (x, y) coordinates to keep the center of the face the same
new_x = max(0, x - int((new_w - w) / 2))
new_y = max(0, y - int((new_h - h) / 2))
# Extract the enlarged face
extracted_face = image[new_y:new_y+new_h, new_x:new_x+new_w]
plt.imshow(extracted_face)
plt.axis('off')
plt.show()
```
